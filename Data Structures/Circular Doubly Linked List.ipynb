### Insertion method to a d linked list

```python
# Create a blank node
# assign a value to it
# assign reference to null
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

# insertion method
# Create head and tail, initialize with null
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    # For iteration print
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            # to break infinite loop
            if node == self.head:
                break
            
    # insertion method
    def insert(self, value, location):
        # create a new node
        new_node = Node(value)
        
        # if head is None -> an empty list
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.prev = new_node
            new_node.next = new_node
            # next/previous is still itself
        
        # if location==head
        elif location==0:
            # 重新建立連結
            self.head.prev = new_node
            self.tail.next = new_node
            new_node.next = self.head
            new_node.prev = self.tail
            self.head = new_node
            
            
        # if location==tail
        elif location==-1:
            # 其實加在 head/加在 tail 是一樣的
            self.tail.next = new_node
            self.head.prev = new_node
            new_node.next = self.head
            new_node.prev = self.tail
            self.tail = new_node
            
            
        # else: location is in between
        else:
            index = 0
            previous_node = self.head
            
            # set previous_node
            while True:
                # let be the previous node
                if index==location-1:
                    break
                # 最後不能超過 tail
                if previous_node == self.tail:
                    break
                previous_node = previous_node.next
                index+=1
            # 如果加在最末端
            if previous_node == self.tail:
                self.tail.next = new_node
                self.head.prev = new_node
                new_node.next = self.head
                new_node.prev = self.tail
                self.tail = new_node
            # 如果非最末端
            else:
                next_node = previous_node.next
                
                previous_node.next = new_node
                next_node.prev = new_node
                new_node.prev = previous_node
                new_node.next = next_node


        
LL = LinkedList()
LL.insert(0,100) 
LL.insert(1,100) 
LL.insert(3,3)
LL.insert(10000,2)

print([node.value for node in LL])

print(LL.tail.next.value)
```

```python

class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

# insertion method
# Create head and tail, initialize with null
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    # For iteration print
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            # to break infinite loop
            if node == self.head:
                break
            
    # insertion method
    def insert(self, value, location):
        # create a new node
        new_node = Node(value)
        
        # if head is None -> an empty list
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.prev = new_node
            new_node.next = new_node
            # next/previous is still itself
        
        # if location==head
        elif location==0:
            # 重新建立連結
            self.head.prev = new_node
            self.tail.next = new_node
            new_node.next = self.head
            new_node.prev = self.tail
            self.head = new_node
            
            
        # if location==tail
        elif location==-1:
            # 其實加在 head/加在 tail 是一樣的
            self.tail.next = new_node
            self.head.prev = new_node
            new_node.next = self.head
            new_node.prev = self.tail
            self.tail = new_node
            
            
        # else: location is in between
        else:
            index = 0
            previous_node = self.head
            
            # set previous_node
            while True:
                # let be the previous node
                if index==location-1:
                    break
                # 最後不能超過 tail
                if previous_node == self.tail:
                    break
                previous_node = previous_node.next
                index+=1
            # 如果加在最末端
            if previous_node == self.tail:
                self.tail.next = new_node
                self.head.prev = new_node
                new_node.next = self.head
                new_node.prev = self.tail
                self.tail = new_node
            # 如果非最末端
            else:
                next_node = previous_node.next
                
                previous_node.next = new_node
                next_node.prev = new_node
                new_node.prev = previous_node
                new_node.next = next_node
    def delete(self, location):
        # if it's empty
        if self.head is None:
            return None
        # if it's only one
        elif self.head == self.tail:
            if location == 0 or -1:
                node = self.head
                
                # 這個指的意思不是 node.prev 那個 node (在這裡就是 node.head自己本身) 把它變成0
                # 而是 node.prev那個項目的 reference
                node.prev = None
                node.next = None
                self.head = None
                self.tail = None
            else:
                return None
        else:
            if location == 0:
                next_node = self.head.next
                next_node.prev = self.tail
                self.tail.next = next_node
                self.head = next_node
                
            elif location == -1:
                prev_node = self.tail.prev
                prev_node.next = self.head
                self.head.prev = prev_node
                self.tail = prev_node
            else:
                index = 0
                prev_node = self.head
                
                while True:
                    if index == location-1:
                        break
                    index+=1
                    prev_node = prev_node.next
                    # 到底了都還沒有找到location，代表超過
                    if prev_node == self.tail:
                        return None
                next_node = prev_node.next.next
                prev_node.next = next_node
                next_node.prev = prev_node
                
                
LL = LinkedList()
LL.insert(0,100) 
LL.insert(1,100) 
LL.insert(3,0) 
LL.insert(4,-1) 
LL.insert(1000,1) 


print([node.value for node in LL])

LL.delete(4)
print([node.value for node in LL])
print(LL.head.value)

```

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

# insertion method
# Create head and tail, initialize with null
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    # For iteration print
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            
    # insertion method
    def insert(self, value, location):
        # create a new node
        new_node = Node(value)
        
        # if head is None -> an empty list
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            # next/previous is still None
        
        # if location==head
        elif location==0:
            self.head.prev = new_node  # 原先 head 的 previous -> new node
            new_node.next = self.head  # new node 的 next -> 原先的 head
            self.head = new_node       # head -> new node
            # tail.next 要重新指定
            
        # if location==tail
        elif location==-1:
            new_node.prev = self.tail  # 將 new node 的 prev 指定為原本的 tail
            self.tail.next = new_node  # 原本 tail node 的下一個指向新node
            self.tail = new_node       # 將 new node 指定為 tail
        
        # else: location is in between
        else:
            index = 0
            previous_node = self.head
            
            # set previous_node
            while True:
                # let be the previous node
                if index==location-1:
                    break
                # 最後不能超過 tail
                if previous_node == self.tail:
                    break
                previous_node = previous_node.next
                index+=1
            # 如果加在最末端
            if previous_node == self.tail:
                self.tail.next = new_node
                new_node.prev = self.tail
                self.tail = new_node
            # 如果非最末端
            else:
                new_node.prev = previous_node       # 建立 new node 的前後關係
                new_node.next = previous_node.next
                previous_node.next.prev = new_node  # 原先下一個node的 prev要連到新 node
                previous_node.next = new_node       # 原先下一個node的

    # 不需要這樣做
    def delete_all(self):
        # 無法像 singly linked list 一樣，只要指定 head/tail to none 就好
        # 因為 node 之間會互相參照
        # erase from head
        current_node = self.head
        while current_node:
            next_node = current_node.next  # 先暫存 next node
            current_node.next = None       # 此 node 的 next 設為 None
            current_node = next_node       # 移動到 next node
        # erase from tail
        current_node = self.tail
        while current_node:
            prev_node = current_node.prev  # 先暫存 next node
            current_node.prev = None       # 此 node 的 prev 設為 None
            current_node = prev_node       # 移動到 prev node

        self.head = None
        self.tail = None
        
    def delete_all2(self):
        # 將 doubly linked 轉為 singly linked，也就是 將 prev -> None
        # 然後將 head 刪掉，之後各個 node因為沒有被 reference，就會被 garbage collector 回收
        current_node = self.head
        while current_node:
            # 轉成 singly linked list
            current_node.prev = None
            current_node = current_node.next
        self.head = None
        self.tail = None

        
LL = LinkedList()
LL.insert(0,100) 
LL.insert(1,100) 
LL.insert(3,0) 
LL.insert(4,-1) 
LL.insert(1000,1) 

print([node.value for node in LL])

LL.delete_all2()
print([node.value for node in LL])
LL.delete_all2()
print([node.value for node in LL])
```
