### Insertion method to a d linked list

```python
# Create a blank node
# assign a value to it
# assign reference to null
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

# insertion method
# Create head and tail, initialize with null
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    # For iteration print
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            
    # insertion method
    def insert(self, value, location):
        # create a new node
        new_node = Node(value)
        
        # if head is None -> an empty list
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            # next/previous is still None
        
        # if location==head
        elif location==0:
            self.head.prev = new_node  # 原先 head 的 previous -> new node
            new_node.next = self.head  # new node 的 next -> 原先的 head
            self.head = new_node       # head -> new node
            # tail.next 要重新指定
            
        # if location==tail
        elif location==-1:
            new_node.prev = self.tail  # 將 new node 的 prev 指定為原本的 tail
            self.tail.next = new_node  # 原本 tail node 的下一個指向新node
            self.tail = new_node       # 將 new node 指定為 tail
        
        # else: location is in between
        else:
            index = 0
            previous_node = self.head
            
            # set previous_node
            while True:
                # let be the previous node
                if index==location-1:
                    break
                # 最後不能超過 tail
                if previous_node == self.tail:
                    break
                previous_node = previous_node.next
                index+=1
            # 如果加在最末端
            if previous_node == self.tail:
                self.tail.next = new_node
                new_node.prev = self.tail
                self.tail = new_node
            # 如果非最末端
            else:
                new_node.prev = previous_node       # 建立 new node 的前後關係
                new_node.next = previous_node.next
                previous_node.next.prev = new_node  # 原先下一個node的 prev要連到新 node
                previous_node.next = new_node       # 原先下一個node的


        
LL = LinkedList()
LL.insert(0,100) 
LL.insert(1,100) 
LL.insert(3,0) 
LL.insert(4,-1) 
LL.insert(1000,1) 

#LL.insert(1,100) 
#LL.insert(1,0)
#LL.insert(2,0)
#LL.insert(3,3)

print([node.value for node in LL])

```

### Deletion of certain node

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

# insertion method
# Create head and tail, initialize with null
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    # For iteration print
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            
    # insertion method
    def insert(self, value, location):
        # create a new node
        new_node = Node(value)
        
        # if head is None -> an empty list
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            # next/previous is still None
        
        # if location==head
        elif location==0:
            self.head.prev = new_node  # 原先 head 的 previous -> new node
            new_node.next = self.head  # new node 的 next -> 原先的 head
            self.head = new_node       # head -> new node
            # tail.next 要重新指定
            
        # if location==tail
        elif location==-1:
            new_node.prev = self.tail  # 將 new node 的 prev 指定為原本的 tail
            self.tail.next = new_node  # 原本 tail node 的下一個指向新node
            self.tail = new_node       # 將 new node 指定為 tail
        
        # else: location is in between
        else:
            index = 0
            previous_node = self.head
            
            # set previous_node
            while True:
                # let be the previous node
                if index==location-1:
                    break
                # 最後不能超過 tail
                if previous_node == self.tail:
                    break
                previous_node = previous_node.next
                index+=1
            # 如果加在最末端
            if previous_node == self.tail:
                self.tail.next = new_node
                new_node.prev = self.tail
                self.tail = new_node
            # 如果非最末端
            else:
                new_node.prev = previous_node       # 建立 new node 的前後關係
                new_node.next = previous_node.next
                previous_node.next.prev = new_node  # 原先下一個node的 prev要連到新 node
                previous_node.next = new_node       # 原先下一個node的

    def delete(self, location):
        # if empty
        if self.head is None:
            return
        
        if location==0:
            # if only one node
            if self.head==self.tail:
                self.head = None
                self.tail = None
            else:
                self.head.next.prev = None
                self.head = self.head.next
        
        elif location==-1:
            # if only one node
            if self.head==self.tail:
                self.head = None
                self.tail = None
            else:
                self.tail.prev.next = None
                self.tail = self.tail.prev
        else:
            index = 0
            previous_node = self.head
            
            while True:
                # 如果 previous node 已經到了 location 的前一位
                if index==location-1:
                    break
                # 如果 previous node 已經到了 tail 的前一位
                if previous_node.next==self.tail:
                    break
                index+=1
                previous_node = previous_node.next
                
            # 如果欲刪除的 location 不存在
            if index!=location-1:
                print('location does not exist')
                return None
            
            # 如果 delete tail
            if previous_node.next==self.tail:
                previous_node.next = None
                self.tail = previous_node
            # 如果 not delet tail
            else:
                next_node = previous_node.next.next
                previous_node.next = next_node
                next_node.prev = previous_node
                

        
LL = LinkedList()
LL.insert(0,100) 
LL.insert(1,100) 
LL.insert(3,0) 
LL.insert(4,-1) 
LL.insert(1000,1) 


print([node.value for node in LL])

LL.delete(0)
print([node.value for node in LL])
print(LL.head.value)

LL.delete(-1)
print([node.value for node in LL])
print(LL.tail.value)

LL.delete(1)
print([node.value for node in LL])
print(LL.tail.value)


```

### Deletion whole linked list

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

# insertion method
# Create head and tail, initialize with null
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    # For iteration print
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            
    # insertion method
    def insert(self, value, location):
        # create a new node
        new_node = Node(value)
        
        # if head is None -> an empty list
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            # next/previous is still None
        
        # if location==head
        elif location==0:
            self.head.prev = new_node  # 原先 head 的 previous -> new node
            new_node.next = self.head  # new node 的 next -> 原先的 head
            self.head = new_node       # head -> new node
            # tail.next 要重新指定
            
        # if location==tail
        elif location==-1:
            new_node.prev = self.tail  # 將 new node 的 prev 指定為原本的 tail
            self.tail.next = new_node  # 原本 tail node 的下一個指向新node
            self.tail = new_node       # 將 new node 指定為 tail
        
        # else: location is in between
        else:
            index = 0
            previous_node = self.head
            
            # set previous_node
            while True:
                # let be the previous node
                if index==location-1:
                    break
                # 最後不能超過 tail
                if previous_node == self.tail:
                    break
                previous_node = previous_node.next
                index+=1
            # 如果加在最末端
            if previous_node == self.tail:
                self.tail.next = new_node
                new_node.prev = self.tail
                self.tail = new_node
            # 如果非最末端
            else:
                new_node.prev = previous_node       # 建立 new node 的前後關係
                new_node.next = previous_node.next
                previous_node.next.prev = new_node  # 原先下一個node的 prev要連到新 node
                previous_node.next = new_node       # 原先下一個node的

    # 不需要這樣做
    def delete_all(self):
        # 無法像 singly linked list 一樣，只要指定 head/tail to none 就好
        # 因為 node 之間會互相參照
        # erase from head
        current_node = self.head
        while current_node:
            next_node = current_node.next  # 先暫存 next node
            current_node.next = None       # 此 node 的 next 設為 None
            current_node = next_node       # 移動到 next node
        # erase from tail
        current_node = self.tail
        while current_node:
            prev_node = current_node.prev  # 先暫存 next node
            current_node.prev = None       # 此 node 的 prev 設為 None
            current_node = prev_node       # 移動到 prev node

        self.head = None
        self.tail = None
        
    def delete_all2(self):
        # 將 doubly linked 轉為 singly linked，也就是 將 prev -> None
        # 然後將 head 刪掉，之後各個 node因為沒有被 reference，就會被 garbage collector 回收
        current_node = self.head
        while current_node:
            # 轉成 singly linked list
            current_node.prev = None
            current_node = current_node.next
        self.head = None
        self.tail = None

        
LL = LinkedList()
LL.insert(0,100) 
LL.insert(1,100) 
LL.insert(3,0) 
LL.insert(4,-1) 
LL.insert(1000,1) 

print([node.value for node in LL])

LL.delete_all2()
print([node.value for node in LL])
LL.delete_all2()
print([node.value for node in LL])
```
