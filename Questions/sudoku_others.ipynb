```python

def find_empty(board):
    for i in range(9):
        for j in range(9):
            if board[i][j]==0:
                return (i,j)
    return None
            
def is_valid(board, num, position):
    r, c = position
    
    # check row
    for i in range(9):
        if board[i][c]==num and i!=c:
            return False

    # check column
    for i in range(9):
        if board[r][i]==num and i!=r:
            return False
        
    # check 3x3
    r_base = r//3*3
    c_base = c//3*3
    for r2 in range(r_base, r_base+3):
        for c2 in range(c_base, c_base+3):
            if board[r2][c2]==num and (r2,c2)!=position:
                return False
    return True
        
def solve(board):
    board =[ [v for v in c] for c in board]
    
    display(board)
    
    find = find_empty(board)
    if not find:
        # found solution
        return True
    else:
        r, c = find
        
    # try number
    for insert_num in range(1,10):
        if is_valid(board, insert_num, (r,c)):
            board[r][c] = insert_num
            
            # 一直 call 下一步，直到找到解為止
            if solve(board):
                return True
            
            # 將數值重置，往下找一個 value
            board[r][c] = 0
            
```
