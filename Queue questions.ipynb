## Three in one
- describe how to use a single Python list to implement three stacks

```python
# method 1
# divide 3 [0,n/3), [n/3, 2*n/3), [2*n/3, n)
# - push  (新增)
# - pop  （剔除並回傳最後一個 element）
# - peek （回傳最後一個 element）

class multipleStack():
    def __init__(self, each_size):
        self.list_numbers = 3
        self.each_size = each_size
        self.items = [None] * self.each_size * self.list_numbers
        self.count = [0] * self.list_numbers  # store how many items now

    def isFull(self, stack_num):
        if self.count[stack_num] == self.each_size:
            return True
        else:
            return False
        
    def isEmpty(self, stack_num):
        if self.count[stack_num] ==0:
            return True
        else:
            return False
        
    def push(self, value, stack_num):
        if self.isFull(stack_num):
            raise Exception('The stack is full')
        else:
            location = (stack_num)*self.each_size + self.count[stack_num]
            self.items[location] = value
            self.count[stack_num] +=1
            
    
    def pop(self, stack_num):
        if self.isEmpty(stack_num):
            raise Exception('The stack is empty')
        else:
            location = (stack_num) * self.each_size + self.count[stack_num] -1
            return_value = self.items[location]
            self.items[location] = None
            self.count[stack_num] -=1
            return return_value
        
    def peek(self, stack_num):
        location = location = (stack_num) * self.each_size + self.count[stack_num] -1
        # to avoid over flow
        if self.isEmpty(stack_num):
            return None
        else:
            return self.items[location]
        
        
m = multipleStack(3)
m.push(1,1)

m.push(2,1)
m.push(3,1)
m.push(4,2)
m.pop(2)
# m.isFull(1)
m.peek(2)
```

## Stack Min
- design a stack, in addition to push and pop, has a function min which returns the minimun element?
- Push, pop, and min should all operate in O(1)

```python
# solution 1
# min()要符合 O(1)
# 要建立一個相應於該 Node之前 的 min，讓刪掉後面 node的時候，可以保留比較用的值
# 當要比較的時候，回傳 self.head 的 value 以及 之前的 min，取小的

class Node():
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev_min = None
    
class Stack():
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.prev_min = value
        else:            
            new_node.next = self.head
            self.head = new_node
            
            # update previous min of head
            next_node = self.head.next
            self.head.prev_min = min(next_node.value, next_node.prev_min)

    def min(self):
        if self.head is not None:
            return min(self.head.value, self.head.prev_min)
        else:
            return None
            
            
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            
    def __str__(self):
        values = [str(n.value) for n in self if n is not None]
        return ' '.join(values)
            
    # 其實 stack 不需要 tail，這樣寫 stack 是多餘的
    def pop(self):
        if self.head is None:
            return None       
        else:
            if self.head == self.tail:
                return_value = self.head.value
                self.head = None
                self.tail = None
            else:
                return_value = self.head.value
                self.head = self.head.next
            return return_value

s = Stack()
s.push(100)
s.push(1)
s.push(2)
s.push(-30)
s.push(3)

print(s)
print(s.min())

s.pop()
print(s)
print(s.min())
s.pop()
print(s)
print(s.min())

s.pop()
print(s)
print(s.min())

s.pop()
print(s)
print(s.min())

s.pop()
print(s)
print(s.min())

```

```python
# solution 2
# 在 Stack 裡面建立另一個 linked list: min_list，紀錄最小 value 的軌跡
# 這兩個 linked list 是不同的，不互相交叉，只是 value 一樣
# 如果 new value<= min_list 最後一個 value，那麼就新增一個 Node，取代原本的 min_list Node
# 這樣如果pop 的時候，其 value 等於 min_list 最後一個 value，那麼就把 min_list 指標往下一個移動
# 小於等於的時候新增 Node 很關鍵，因為有時候 node為同值
# 其實平常建立新的 node, 就跟這個min_list 一樣，不一定需要一個 new_node 做緩衝，可以直接 assign

# 注意 node 的寫法，可以指定 value, next
class Node():
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next
    
# stack 其實不需要 tail
class Stack():
    def __init__(self):
        self.top = None
        # self.tail = None
        self.min_list = None
        
    def push(self, value):
        new_node = Node(value)
        if self.top is None:
            self.top = new_node
            self.min_list = Node(value=value)
        else:            
            new_node.next = self.top
            self.top = new_node
            
            # update previous min of head
            #if self.min.value<=
            if self.top.value<=self.min_list.value:
                new_min_node = Node(value=value, next=self.min_list)
                self.min_list = new_min_node

    def min(self):
        if self.min_list is not None:
            return self.min_list.value
        else:
            return None
            
            
    def __iter__(self):
        node = self.top
        while node:
            yield node
            node = node.next
            
    def __str__(self):
        values = [str(n.value) for n in self if n is not None]
        return ' '.join(values)
            
    def pop(self):
        if self.top is None:
            return None       
        else:
            return_value = self.top.value
            self.top = self.top.next
                
            # update min_list
            if return_value==self.min_list.value:
                self.min_list = self.min_list.next
            
            return return_value
        
s = Stack()
s.push(-10)
s.push(1)
s.push(-22)
s.push(-1)
s.push(-22)
print(s)
print(s.min())

s.pop()
print(s)
print(s.min())

s.pop()
print(s)
print(s.min())

s.pop()
print(s)
print(s.min())

s.pop()
print(s)
print(s.min())
```

## Stack of Plates
- if the stack gets too high, it might topple
- start a new stack when previous stack exceed some threshold
- To implement a data structure SetOfStacks that mimic this
- SetOfStacks 要包含多個 stacks，如果stack超過一定數量
- Pop() 使用方式也要一樣
- 新增一個 popAt(N)，讓我們可以 取出第N個 subplate 的 項目

```python
# method 1
# Stack of plates
# 在 top, head 之間，很容易搞混，要小心
# use doubly linked list
# very complicated，但其實題目沒有要我們用自己建立的 linked list

class Node():
    def __init__(self,value=None):
        self.value = value
        self.next = None
        
class Subplate():
    # use doubly linked list to manage subplates
    def __init__(self):
        self.top = None
        self.next_subplate = None
        self.prev_subplate = None
        self.num = 0
        
    def push(self, value):
        new_node = Node(value)
        if self.top is None:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.num +=1
            
    def pop(self):
        if self.top is None:
            return None
        else:
            return_value = self.top.value
            self.top = self.top.next
            self.num -=1
            return return_value
        
    def peek(self):
        if self.top is None:
            return None
        else:
            return self.top.value
        
    def __iter__(self):
        node = self.top
        while node:
            yield node
            node = node.next
    
    def __str__(self):
        values = [str(n.value) for n in self if n is not None]
        return ' '.join(values)
        
        
class Plates():
    def __init__(self,limit):
        self.limit = limit
        self.head = None
        self.tail = None
        
    def __iter__(self):
        subplate = self.head
        while subplate:
            yield subplate
            subplate = subplate.next_subplate
            
    def __str__(self):
        # return all plates, call the str(p), get representation of subplates, then combine together
        all_subplate = [str(p) for p in self if p is not None]
        return '\n'.join(all_subplate)
        
    def push(self, value):
        if self.head is None:
            new_subplate = Subplate()
            new_subplate.push(value)
            self.head = new_subplate
            self.tail = new_subplate
            
        # if full
        elif self.head.num==self.limit:
            new_subplate = Subplate()
            new_subplate.push(value)
            new_subplate.next_subplate = self.head
            self.head.prev_subplate = new_subplate
            self.head = new_subplate
        else:
            self.head.push(value)
    
    def popAt(self, num):
        subplate = self.head
        
        index = 0
        while True:
            if subplate is None:
                raise Exception('No such plate')
            elif index==num:
                return_value = subplate.pop()
                
                if subplate.num == 0:
                    # at top plate
                    if num ==0:
                        # if it is only one plate left, then clear
                        if self.head == self.tail:
                            self.tail = None
                            self.head = None
                        # if it is not tail
                        else:
                            next_plate = self.head.next_subplate
                            next_plate.prev_subplate = None
                            self.head.next_subplate = None
                            self.head = next_plate
                    else:
                        if subplate == self.tail:
                            prev_plate = subplate.prev_subplate
                            next_plate = subplate.next_subplate
                            prev_plate.next_subplate = next_plate
                            self.tail = prev_plate
                        else:
                            prev_plate = subplate.prev_subplate
                            next_plate = subplate.next_subplate
                            prev_plate.next_subplate = next_plate
                            next_plate.prev_subplate = prev_plate                        
                return return_value
            index +=1
            subplate = subplate.next_subplate
        
    def pop(self):
        if self.head is None:
            return None
        else:
            return_value = self.head.pop() 
            # if subplate is empty -> delete it
            if self.head.num == 0:
                next_plate = self.head.next_subplate
                next_plate.prev_subplate = None
                self.head.next_subplate = None
                self.head = next_plate
    

p = Plates(limit=3)
p.push(1)
p.push(2)
p.push(3)
p.push(4)
p.push(5)
p.popAt(1)
p.popAt(1)
p.popAt(1)
p.popAt(0)
p.popAt(0)
p.push(5)
print(p)
```

```python
# method 2 use Python link
# much more easier 
# 題目沒有要我們用困難的 doubly linked list，不必自己升級！！！

class Plates():
    def __init__(self, capacity):
        # create items stack
        self.items = []
        self.capacity = capacity
        
    def push(self, value):
        if self.items ==[]:
            self.items.append([])
        if len(self.items[0])==self.capacity:
            self.items.insert(0,[])
        self.items[0].insert(0,value)
        
    def pop(self):
        if self.items==[]:
            return None
        else:
            return_value = self.items[0].pop(0)
            if self.items[0] ==[]:
                del self.items[0]
            return return_value
        
    def popAt(self, num):
        if self.items==[]:
            return None
        else:
            return_value = self.items[num].pop(0)
            if self.items[num] ==[]:
                del self.items[num]
            return return_value

    def __str__(self):
        #return '\n'.join([' '.join(map(str, v)) for v in self.items if v!=[]])
        return_list = []
        for plate in self.items:
            string = [str(v) for v in plate]
            return_list.append(' '.join(string))
        return '\n'.join(return_list)
    
p = Plates(3)
p.push(1)
p.push(2)
p.push(3)
p.push(9)
p.popAt(1)
p.popAt(1)
p.popAt(1)
p.popAt(1)

#

print(p)
```

```python
# method 0: 使用 linked list,紀錄盤的數量，print的時候注意plate size
# 這不是答案要的

class Node():
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next
        
class Plate():
    def __init__(self, limit):
        self.top = None
        self.plate_num = 0
        self.limit = limit
        
    def push(self, value):
        new_node = Node(value)
        if self.top is None:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.plate_num = self.plate_num+1
        
    def pop_at(self, location):
        node = self.top
        
        if location==0:
            if node:
                return_value = node.value
                self.top = self.top.next
                self.plate_num = self.plate_num-1
                return return_value
        else:
            # to shift the node to the node before deletion
            # locate 前一個 location
            for index in range(location-1):
                # 如果 run 完了
                if node.next is None:
                    print('No index is found')
                    return None
                node = node.next
            return_value = node.next.value
            node.next = node.next.next
            self.plate_num = self.plate_num-1

            return return_value
            
    def pop(self):
        if self.top is None:
            return None
        else:
            return_value = self.top.value
            self.top = self.top.next            
            self.plate_num = self.plate_num-1
            return return_value
        
        
        
    def __iter__(self):
        node = self.top
        while node:
            yield node
            node = node.next
        
    def __str__(self):
        if self.top is None:
            return 'The plate is empty'
        else:
            return_str = []
            node = self.top
            for i in range(self.plate_num%self.limit):
                return_str.append(str(node.value))
                node = node.next

            count = 0
            while node:
                if count==0:
                    return_str.append('\n')
                return_str.append(str(node.value))
                count = (count+1)%self.limit
                node = node.next

            return ''.join(return_str if return_str[0]!='\n' else return_str[1:])

p = Plate(3)
p.push(1)
p.push(2)
p.push(3)
p.push(4)
p.push(5)

print(p)

p.pop_at(4)
print(p)
p.pop_at(3)
print(p)
```
